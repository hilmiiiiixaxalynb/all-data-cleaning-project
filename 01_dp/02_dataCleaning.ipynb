```python
import pandas as pd 
```

```python
df = pd.read_csv(r'C:\Users\hilmi\Documents\coding\pandas latihan analisis\data\my_file (1).csv')
```

1. inspecting dirty values

```python

df.info()
df.isnull().sum()
df.duplicated().sum()
df.describe(include='all')
```

2. fixing non breaking space

```python
df.columns = df.columns.str.replace('\xa0', ' ')
```

3. cleaning data

```python
df['Peak'] = df['Peak'].str.replace(r'\[|\]', '', regex=True)
df['All Time Peak'] = df['All Time Peak'].str.replace(r'\[|\]', '', regex=True)
df['Peak'] = df['Peak'].str.replace('c', '', regex=True)
df['Actual gross'] = df['Actual gross'].replace(r'\[|\]', '', regex=True)
df['Actual gross'] = df['Actual gross'].replace('[a-z]', '', regex=True)
df['Tour title'] = df['Tour title'].replace(r'[†‡\+\#\*\^\[\]]', '', regex=True)
df['Ref.'] = df['Ref.'].replace(r'\[|\]', '', regex=True)
df['Ref.'] = df['Ref.'].replace('d', 'Unknown', regex=True)
df.iloc[13, 6] = df.iloc[13, 6].replace('Living Proof: The Farewell Tour 21a', 'Living Proof: The Farewell Tour')
df.iloc[2, 6] = df.iloc[13, 6].replace('Sticky & Sweet Tour 4a', 'Sticky & Sweet Tour')

# missing values
df['Peak'] = df['Peak'].fillna('Unknown')
df['All Time Peak'] = df['All Time Peak'].fillna('Unknown')
df.sample(4)
```

4. changing data type for analysis

```python
df['Actual gross'] = df['Actual gross'].astype(str).replace(r'[^0-9.]', '', regex=True).astype(float)
df['Adjusted gross (in 2022 dollars)'] = df['Adjusted gross (in 2022 dollars)'].astype(str).replace(r'[^0-9.]', '', regex=True).astype(float)
df['Average gross'] = df['Actual gross'].astype(str).replace(r'[^0-9.]', '', regex=True).astype(float)
df
```
